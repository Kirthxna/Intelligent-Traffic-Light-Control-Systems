```python
import matplotlib.pyplot as plt

with open("epsilon_values.txt", "r") as f:
	lines = f.read()
 
lines = lines.splitlines()
dqn_epsilon_values = []

for line in lines:
    dqn_epsilon_values.append(float(line))

fig, ax = plt.subplots()

ax.plot(list(range(len(dqn_epsilon_values))), dqn_epsilon_values)
ax.set_xlabel('Epsilon')
ax.set_ylabel('Episode')
ax.set_title('Epsilon decay graph')

plt.show()
```
